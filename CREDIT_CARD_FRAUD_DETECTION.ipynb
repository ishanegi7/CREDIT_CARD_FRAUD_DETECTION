```python
#IMPORTING THE DEPENDENCIES
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


```

```python
#loading tha data to pd dataframe
credit_card_data= pd.read_csv('/creditcard.csv')
```

```python
#5 lines of data
credit_card_data.head()
#data info
credit_card_data.info()
```

```python
#no of missing values in each column

credit_card_data.isnull().sum()
```

```python
#to see distribution of legit transaction and  fraudalent transaction
credit_card_data['Class'].value_counts()
```

THE DATA IS HIGHLY UNBALANCED


0--> NORMAL TRANSACTION

1--> FRAUDALENT TRANSACTION


```python
#separating the data for analaysis
LEGIT= credit_card_data[credit_card_data.Class ==0]
FRAUD= credit_card_data[credit_card_data.Class ==1]
```

```python
print(LEGIT.shape)
print(FRAUD.shape)
```

```python
#statistical measure of the data
LEGIT.Amount.describe()
```

```python
FRAUD.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

UNDER SAMPLING


Build a sample dataset containing similar distribtution of normal transaction and fraudalent transaction..


```python
LEGIT_sample= LEGIT.sample(n=492)
```

CONCATING TWO DF


```python
LEGIT_sample= LEGIT.sample(n=492)
new_dataset =pd.concat([LEGIT_sample , FRAUD ],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Tarrgets


```python
X = new_dataset.drop('Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing data


```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , stratify=Y , random_state=2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

MODEL TRAINING


Logistic Regression

```python
#training the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train , Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data :', training_data_accuracy)
```

```python
#accuracy on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data :', test_data_accuracy)
```
